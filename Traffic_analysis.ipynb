```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/traffic.csv")

# Convert 'Time' column to datetime format for better analysis
df["Time"] = pd.to_datetime(df["Time"], format="%I:%M:%S %p").dt.hour  # Extract only the hour

# Ensure 'Traffic Situation' is categorical for proper visualization
df["Traffic Situation"] = df["Traffic Situation"].str.lower().astype("category")

# Set figure size
plt.figure(figsize=(14, 6))

# Task 1 Vehicle Distribution Analysis:

# Boxplot for vehicle counts
plt.subplot(1, 2, 1)
sns.boxplot(data=df[["CarCount", "BikeCount", "BusCount", "TruckCount"]])
plt.title("Boxplot of Vehicle Counts")
plt.ylabel("Count")


```

```python
# Histogram for vehicle counts
df[["CarCount", "BikeCount", "BusCount", "TruckCount"]].plot.hist(alpha=0.5, bins=20, figsize=(14, 6))
plt.title("Histogram of Vehicle Counts")
plt.xlabel("Count")

plt.show()
```

```python
# task 2: Traffic Situation Distribution:
# Bar chart
plt.figure(figsize=(12, 5))
sns.countplot(x=df["Traffic Situation"], order=df["Traffic Situation"].value_counts().index, palette="viridis")
plt.title("Distribution of Traffic Situations")
plt.xlabel("Traffic Situation")
plt.ylabel("Count")
plt.show()

# Pie chart
df["Traffic Situation"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(8, 8), colors=["red", "orange", "yellow", "green"])
plt.title("Traffic Situation Distribution")
plt.ylabel("")
plt.show()
```

```python
# task 3: Variation by Day of the Week:
plt.figure(figsize=(14, 8))

# Bar plot for each vehicle type
plt.subplot(2, 2, 1)
sns.barplot(x="Day of the week", y="CarCount", data=df, estimator=sum, ci=None, palette="Blues")
plt.xticks(rotation=45)
plt.title("Car Count by Day of the Week")
plt.ylabel("Car Count")

plt.subplot(2, 2, 2)
sns.barplot(x="Day of the week", y="BikeCount", data=df, estimator=sum, ci=None, palette="Greens")
plt.xticks(rotation=45)
plt.title("Bike Count by Day of the Week")
plt.ylabel("Bike Count")

plt.subplot(2, 2, 3)
sns.barplot(x="Day of the week", y="BusCount", data=df, estimator=sum, ci=None, palette="Oranges")
plt.xticks(rotation=45)
plt.title("Bus Count by Day of the Week")
plt.ylabel("Bus Count")

plt.subplot(2, 2, 4)
sns.barplot(x="Day of the week", y="TruckCount", data=df, estimator=sum, ci=None, palette="Reds")
plt.xticks(rotation=45)
plt.title("Truck Count by Day of the Week")
plt.ylabel("Truck Count")

plt.tight_layout()
plt.show()
```

```python
# 4-7. Vehicle Counts vs. Traffic Situation (Scatter, Box, Violin Plots)
# Define function for visualization
def plot_vehicle_vs_traffic(vehicle_column, title):
    plt.figure(figsize=(12, 5))

    # Scatter plot
    plt.subplot(1, 3, 1)
    sns.scatterplot(x=df[vehicle_column], y=df["Traffic Situation"], hue=df["Traffic Situation"], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    # Box plot
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df["Traffic Situation"], y=df[vehicle_column], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    # Violin plot
    plt.subplot(1, 3, 3)
    sns.violinplot(x=df["Traffic Situation"], y=df[vehicle_column], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    plt.show()

# Call function for each vehicle type
plot_vehicle_vs_traffic("CarCount", "Car Count")
plot_vehicle_vs_traffic("BikeCount", "Bike Count")
plot_vehicle_vs_traffic("BusCount", "Bus Count")
plot_vehicle_vs_traffic("TruckCount", "Truck Count")
```

```python
# 8. Total Vehicle and Count Traffic Situation
# Group by 'Traffic Situation' and sum for each vehicle type
vehicle_types = ["CarCount", "BikeCount", "BusCount", "TruckCount"]

vehicle_totals = df.groupby('Traffic Situation')[vehicle_types].sum()

# Plot the results
vehicle_totals.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Total Vehicle Count by Traffic Situation for Each Vehicle Type')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Vehicle Types')
plt.show()
```

```python
# Sum the vehicle counts for each row (total vehicle count)
df['Total'] = df[vehicle_types].sum(axis=1)

# Group by 'Traffic Situation' and sum the total count for each traffic situation
traffic_totals = df.groupby('Traffic Situation')['Total'].sum()

# Plot the total vehicle count for each traffic situation
plt.figure(figsize=(8, 6))
traffic_totals.plot(kind='bar', color='coral')
plt.title('Total Vehicle Count by Traffic Situation')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Heavy', 'High', 'Normal', 'Low'], rotation=0)  # Set custom x-axis labels
plt.show()
```

```python
# Task 9: Bar chart of total vehicle count per hour
plt.figure(figsize=(12, 6))
sns.barplot(x=df["Time"], y=df["Total"], estimator=sum, ci=None, palette="magma")
plt.title("Total Vehicle Count by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Vehicles")
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("traffic.csv")

# Convert 'Time' to datetime format
df['Time'] = pd.to_datetime(df['Time'], format='%I:%M:%S %p').dt.hour
```

#**Traffic Situation Distribution by Day of the Week**

**"Heavy" traffic is more common on weekdays, particularly on Friday.**

**"Normal" traffic dominates on weekends, with occasional "high" traffic peaks.**

**"Low" traffic levels are relatively stable across all days, mostly appearing during non-peak hours.**


```python
#10 Traffic situation distribution by day of the week (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Day of the week', y='Total', hue='Traffic Situation', order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], palette="Set2")
plt.title("Traffic Situation Distribution by Day of the Week", fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Total Vehicle Count", fontsize=12)
plt.legend(title="Traffic Situation", title_fontsize='13', fontsize='11')
plt.xticks(rotation=45, fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#10 (another way) Traffic situation distribution by day of the week
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Day of the week', hue='Traffic Situation', order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Traffic Situation Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Count")
plt.legend(title="Traffic Situation")
plt.xticks(rotation=45)
plt.show()
```

#**Correlations between different vehicle types**

**Strong Positive Correlations:**

CarCount & BikeCount (0.72), CarCount & BusCount (0.69), and BikeCount & BusCount (0.57) show that these vehicle types tend to increase together, indicating shared usage patterns or similar traffic conditions.

**Negative Correlation with TruckCount:**

TruckCount has a negative correlation with CarCount (-0.63), BikeCount (-0.58), and BusCount (-0.55), suggesting that areas with high truck traffic have fewer other vehicles, possibly due to road type or traffic regulations.

**When there are more trucks, there tend to be fewer cars, bikes, and buses, possibly because trucks operate in different places.**

```python
#11 Correlations between different vehicle types
plt.figure(figsize=(8, 6))
sns.heatmap(df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Between Vehicle Types")
plt.show()
```

#**Traffic Situation Distribution by Hour**

**Traffic is mostly "low" or "normal" during late night and early morning hours.**

**"Heavy" traffic occurs predominantly between 6-10 AM and 3-7 PM, aligning with rush hours.**

**"High" traffic is scattered throughout mid-day and evening but is less frequent than "heavy" traffic.**

```python
#12 Traffic situation distribution by hour (Box Plot)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Time', y='Total', hue='Traffic Situation', palette="Set1")
plt.title("Traffic Situation Distribution by Hour", fontsize=14, fontweight='bold')
plt.xlabel("Hour of the Day", fontsize=12)
plt.ylabel("Total Vehicle Count", fontsize=12)
plt.legend(title="Traffic Situation", title_fontsize='13', fontsize='11')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#12 (Another way) Traffic situation distribution by hour
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Time', hue='Traffic Situation')
plt.title("Traffic Situation Distribution by Hour")
plt.xlabel("Hour of the Day")
plt.ylabel("Count")
plt.legend(title="Traffic Situation")
plt.show()
```

#**Total Vehicle Count Distribution by Day of the Week**

**The boxes for all days look quite similar, meaning the total vehicle count remains fairly consistent throughout the week.**

**Friday has the most variation (larger box and longer whiskers), suggesting that traffic conditions on Fridays are less predictable.**


```python
#13 Distribution of total vehicle counts for each day of the week
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Day of the week', y='Total', order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Total Vehicle Count Distribution by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Total Vehicle Count")
plt.xticks(rotation=45)
plt.show()
```

#**Variance in Vehicle Counts Across Types**

**Car count has the highest variance, meaning significant fluctuations throughout the day.**

**Bus and bike counts have moderate variance.**

**Truck count shows the least variance, indicating a steady flow.**

```python
#14 Variance in vehicle counts across vehicle types
plt.figure(figsize=(8, 6))
df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].var().plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.title("Variance in Vehicle Counts Across Types")
plt.ylabel("Variance")
plt.show()
```

#**Average Vehicle Count Over Time**

**Car count peaks during morning (around 6-9 AM) and evening rush hours (around 4-7 PM).**

**Other vehicle types (bikes, buses, and trucks) remain relatively stable throughout the day.**

**Bus and bike counts increase slightly during peak hours, while truck count remains fairly constant.**

```python
#15 Average vehicle count for each type over time
df.groupby('Time')[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].mean().plot(figsize=(12, 6))
plt.title("Average Vehicle Count Over Time")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Count")
plt.legend(title="Vehicle Type")
plt.show()
```
